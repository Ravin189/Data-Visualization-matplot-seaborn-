# Matplotlib Python

```python
# line chart
```

```python
plt.plot(file['price'],color='y')
plt.title("Hotel price")
plt.show
```

```python
#Bar Plot
```

```python
x= file['minimum_nights']
y= file['number_of_reviews']
c=('g','r')

plt.bar(x,y,color=c)
plt.title("reviews by nights")
plt.show
```

```python
#Scatter Plot
```

```python
x1= file['latitude']
y1= file['longitude']

plt.scatter(x1,y1)
plt.title('latitude & longitude')
plt.show
```

```python
# stem plot
```

```python
x2=file['host_id']
y2 =file['price']

plt.title('price')
plt.stem(x2,y2)
plt.show
```

# Seaborn

```python
#line plot
```

```python
x3=file['host_id']
y3 = file['price']

sns.lineplot(x3,y3,data=file,hue='neighbourhood_group')
sns.show
```

```python
#boxplot
```

```python
a=file['room_type']
b=file['price']

sns.boxplot(a,b,data=file,hue='neighbourhood_group')
sns.show
```

```python
#pair plot
```

```python
sns.pairplot(file,size=3,hue='host_name')
sns.show
```

```python
#violin plot
```

```python
sns.violinplot(xv='latitude',yv='longitude',data=file)
sns.show
```

# HrDataset File

```python
filehr =pd.read_csv('Hr Dataset.csv')
```

```python
filehr.head()
```

```python
filehr.isnull().sum()
```

# Matplotlib Python-HrData

```python
# line chart
```

```python
plt.plot(filehr['Salary'],color='y')
plt.title('Salary chart')
plt.show
```

```python
#bar plot
```

```python
h1=filehr['EmpID']
h2= filehr['EmpSatisfaction']
c=('y','r')
plt.title('Performance')
plt.bar(h1,h2,color=c)
plt.show
```

```python
#Line plot
```

```python
plt.plot(filehr['Absences'])
plt.title('Absences Report')
plt.show
```

```python
#Violin Plot
```

```python
plt.violinplot(filehr['EmpSatisfaction'])
plt.grid()
plt.title('Emp Rating Ratio')
plt.show
```

```python
#Scatter Plot
```

```python
s1=filehr['EngagementSurvey']
s2=filehr['EmpID']
plt.title('Emp Survey')

plt.scatter(s1,s2)
plt.show
```

# Seaborn-Hr Data

```python
#Box Plot
```

```python
sns.boxplot(x=filehr['PerformanceScore'],y=filehr['Salary'],data=filehr).set_title('Performance by salary')
sns.show
```

```python
#Point Plot
```

```python

sns.pointplot(x=filehr['Absences'],y=filehr['EmpID'],data=filehr)
sns.show
```

```python
#Lmplot
```

```python
sns.lmplot(x='EmpID',y='Salary',data=filehr,fit_reg=False)
sns.show
```

```python
# Violin plot
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='EmpID',y='RecruitmentSource',data=filehr)
sns.show
```
